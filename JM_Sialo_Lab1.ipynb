```python
import matplotlib.pyplot as plt
import io
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/My Drive/AI-ML/day.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data['season_label'] = data['season'].map({1:'winter',2:'spring',3:'fall',4:'summer'})
data['yr_label'] = data['yr'].map({0:'2011',1:'2012'})
data['holiday_label'] = data['holiday'].map({0:'saturday',1:'sunday'})
data['mnth_label'] = data['mnth'].map({1:'january',2:'feburary',3:'march',4:'april',5:'may',6:'june',7:'july',8:'august',9:'september',10:'october',11:'november',12:'december'})
data['weekday_label'] = data['weekday'].map({0:'sunday',1:'monday',2:'tuesday',3:'wednesdayday',4:'thrusday',5:'friday',6:'saturday'})
data['workingday_label'] = data['workingday'].map({0:'holiday',1:'work-day'})
data['weathersit_label'] = data['weathersit'].map({1:'sunny',2:'cloudy',3:'windy',4:'rainy'})

data.head()
```

```python
data.corr()
```

```python
day_info=data[['temp','hum','windspeed','casual','workingday']]
sns.pairplot(day_info,corner=True,hue="workingday",diag_kind='hist')
```

```python
plt.title('Record distribution by count of casual users')
sns.histplot(data=data, x='casual',color='darkorange',kde=True)
plt.axvline(data['casual'].mean(),color='red',linestyle='--',linewidth=3)
plt.xlabel('CasualUsers')
plt.ylabel('records')
plt.show()
```

```python
plt.title('Record distribution by count of registered users')
sns.histplot(data=data, x='registered',color='limegreen',kde=True)
plt.axvline(data['registered'].mean(),color='red',linestyle='--',linewidth=3)
plt.xlabel('RegisteredUsers')
plt.ylabel('records')
plt.show()
```

```python
sns.countplot(data=data,x='weathersit_label',palette='gist_earth')
plt.xlabel('weather')
plt.ylabel('average')
plt.show()
```

```python
sns.stripplot(data=data, x='yr_label',y='cnt',palette='Set2')

plt.xlabel('years')
plt.ylabel('average')
plt.show()
```

```python
sns.swarmplot(data=data, x='yr_label',y='cnt',palette='ocean')

plt.xlabel('years')
plt.ylabel('average')
plt.show()
```

```python
sns.scatterplot(data=data,x='casual',y='registered',hue='yr_label')

plt.xlabel('casual users',fontsize=10)
plt.xticks(color='navy',fontsize=11)

plt.ylabel('registered users',fontsize=10)
plt.yticks(color='navy',fontsize=11)

plt.legend(loc='upper right')
plt.show()
```
